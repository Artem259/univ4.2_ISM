# <span style="color:#FF5733">Грузія</span>

<div style="border:2px solid #FF5733; padding:5px; display:block">

**[Гру́зія](https://uk.wikipedia.org/wiki/%D0%93%D1%80%D1%83%D0%B7%D1%96%D1%8F)** — держава у Східній Європі, на південно-східному узбережжі Чорного моря. Розташована на перехресті Європи й Азії. Має спільні кордони з такими країнами: на півночі — Росія, на півдні — Туреччина і Вірменія, на сході — Азербайджан. Грузія є членом Ради Європи, Євроконтролю, Організації з безпеки і співробітництва в Європі, і Організації за демократію та економічний розвиток.

Грузія є *унітарною, напівпрезидентською республікою*, з урядом, обраним за принципом *представницької демократії*.

### <span style="color:green">Історія та Культура</span>
- #### Історія
  - Грузія володіє багатою історією, налічуючи тисячоліття свого існування. У середньовіччі, Грузія відігравала ключову роль у торгівлі та культурному обміні на Шовковому шляху.
  - З 19 століття, Грузія стала частиною Російської імперії, а пізніше - СРСР. Незалежність була проголошена в 1991 році після розпаду СРСР.
- #### Мова та Писемність
  - Грузинська мова використовує власний алфавіт, відомий як *мхедрулі*. Це один з найдавніших письмових алфавітів у світі, створений в 3-5 століттях.
- #### Традиції та Обряди
  - Грузинська культура відзначається унікальністю та гостинністю. Традиційні танці та музика є важливою частиною спадщини, а святкування релігійних та національних свят - необхідною частиною життя.

### <span style="color:green">Економіка та Туризм</span>
- #### Економіка
  - Грузія має розвинену економіку, зокрема сільське господарство, туризм та легку промисловість. Виноробство також грає ключову роль у її господарстві.
- #### Туризм
  - Туризм в Грузії стає все популярнішим завдяки привабливим природним пейзажам та історичним пам'яткам. Старовинні міста, такі як Тбілісі та Мцхета, привертають увагу туристів своєю архітектурою та історією.

### <span style="color:green">Видатні особистості</span>
- #### Політики
   - *Міхеїл Саакашвілі:* бувший президент Грузії, відомий своєю реформаторською діяльністю.
- #### Культурні діячі
   - *Нино Ціцішвілі:* визначна грузинська письменниця, авторка "Книги джентльмена".
   - *Шота Руставелі:* середньовічний грузинський поет, автор "Витязь у тигровій шкурі".

### <span style="color:green">Пам'ятки культури та природи</span>
- #### Архітектура
   - У Грузії розташовані численні церкви, зокрема відома *Светіцховели* в Мцхеті.
- #### Природа
   - *Гори Кавказу:* Грузія славиться своїми величними горами, частиною Кавказьких гір.

### <span style="color:green">Гастрономія</span>
- #### Національна кухня
  - Грузинська кухня славиться своєю смакотою та різноманіттям. Кенгалі, хачапурі, хінкалі та сатсіві - лише декілька прикладів вишуканих страв.
- #### Виноробство
  - Виробництво вина в Грузії має довгу історію та визнану якість.
